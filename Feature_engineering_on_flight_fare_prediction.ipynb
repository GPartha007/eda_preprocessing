```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("Flight_Fare.xlsx")
```

```python
df.head()
```

```python
# Treating Null values
df.isnull().sum()
```

```python
# Since there only 1 null values in Route & Total_Stops, we dropped that row:
df = df.dropna()
```

```python
# Treating Date_of_Journey feature:
# Creating a column "Journey_day" & "Journey_month", dropping "Date_of_Journey"

date = df["Date_of_Journey"]

new_day_arr = []
new_month_arr = []

for i in date:
    if len(i) == 10:
        new_day_arr.append(i[0:2])
        new_month_arr.append(i[3:5])
    if len(i) == 9:
        new_day_arr.append(i[0:1])
        new_month_arr.append(i[2:4])
        
df["Journey_day"] = new_day_arr
df["Journey_month"] = new_month_arr

# Dropping "Date_of_Journey"
df = df.drop(columns=["Date_of_Journey"], axis=1)
```

```python
# Treating Dep_Time feature:
# Creating a column "Dep_hour" & "Dep_min", dropping "Dep_Time"

dep_time = df["Dep_Time"]

new_hour_arr = []
new_min_arr = []

for x in dep_time:
    new_hour_arr.append(x[:2])
    new_min_arr.append(x[3:])

df["Dep_hour"] = new_hour_arr
df["Dep_min"] = new_min_arr

# Dropping "Date_of_Journey"
df = df.drop(columns=["Dep_Time"], axis=1)
```

```python
# Treating Arrival_Time feature:
# Creating a column "Arrival_hour" & "Arrival_min", dropping "Arrival_Time"

arv_time = df["Arrival_Time"]

new_arv_hour_arr = []
new_arv_min_arr = []

for j in arv_time:
    new_arv_hour_arr.append(j[:2])
    new_arv_min_arr.append(j[3:5])

df["Arrival_hour"] = new_arv_hour_arr
df["Arrival_min"] = new_arv_min_arr

# Dropping "Date_of_Journey"
df = df.drop(columns=["Arrival_Time"], axis=1)
```

```python
# Treating Duration feature:
# Evaluating hrs as 60 mins

df["Duration"] = df["Duration"].str.replace("h", "*60").str.replace(" ", "+").str.replace("m", "*1").apply(eval)
```
